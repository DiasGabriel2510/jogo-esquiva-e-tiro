```python
#Importar as bibliotecas
import pygame
import random

pygame.init()

#Define as dimensoes da tela do jogo
largura, altura = 800, 600
tela = pygame.display.set_mode((largura, altura))

#Definir titulo do jogo
pygame.display.set_caption('Spaceship game')

#selecionando imagem da espaconave e definindo tamanho
jogador = pygame.image.load('spaceship.png')
jogador = pygame.transform.scale(jogador, (50, 50))

#selecionando imagem do inimigo e definindo tamanho
imagem_inimigo = pygame.image.load('enemy.png')
imagem_inimigo = pygame.transform.scale(imagem_inimigo, (50, 50))

#selecionando imagem da explosao e definindo tamanho
imagem_explosao = pygame.image.load('explosion.png')
imagem_explosao = pygame.transform.scale(imagem_explosao, (50, 50)) 

#Lista de inimigos em explosao
explosoes = []

#selecionando som do tiro
som_tiro = pygame.mixer.Sound('shot.mp3')

#Definicao de duas cores em RGB
preto = (0, 0 , 0)
branco = (255, 255, 255)

#Configuracoes do jogador no jogo

#tamanho
tamanho_jogador = 50

#posicao do jogador definida horizontalmente no centro da tela e verticalmente acima da parte inferior
#velocidade do jogador a cada frame
posicao_jogador = [largura // 2, altura - 2 * tamanho_jogador]
velocidade_jogador = 6

#Configuracoes do inimigo no jogo

#quantidade de inimigos que aparecem
#tamanho do inimigo proporcional ao do jogador
num_inimigos = 15
tamanho_inimigo = 50

largura = 800 #largura janela de exibicao

#Iniciando lista de inimigos vazia
posicoes_inimigos = []

#Loop "for" para adicionar a cada inimigo na lista
for _ in range(num_inimigos):
    inimigo = {
        'x': random.randint(0, largura - tamanho_inimigo),
        'y': random.randint(-200, 0,),
        'velocidade': random.randint(3, 7)
        #Cria um dicionario para o inimigo com uma posicao 'x', 'y' e velocidade aleatoria
        #largura - tamanho_inimigo e usada para garantir que ele nao apareca parcialmente fora da tela
    }
    
    #Adiciona o dicionario do inimigo na lista 'posicaoes_inimigos'
    posicoes_inimigos.append(inimigo)

#Configuracoes para os tiros disparados
tiros = []
velocidade_tiro = 7

#Variaveis da pontuacao (inicia com zero) e quantidade de vidas (inicia com tres)
pontos = 0
vidas = 5

#Configuracao do texto exibido na tela
fonte = pygame.font.SysFont(None, 36)

#Inicio do loop principal do jogo

#Objeto Clock usado para controlar o tempo dentro do jogo (limita taxa de quadros)
relogio = pygame.time.Clock()

#Variavel que define fim do jogo
fim_de_jogo = False

#Carrega e dimensiona a imagem, cobrindo toda a tela
imagem_fundo = pygame.image.load('space.jpg')
imagem_fundo = pygame.transform.scale(imagem_fundo, (largura, altura))

#Funcao 'criar_novo_inimigo'
def criar_novo_inimigo():
    novo_inimigo={
        'x': random.randint(0, largura - tamanho_inimigo),
        'y': random.randint(-200, 0),
        'velocidade': random.randint(3, 7)
        #Novamente cria um dicionario para o novo_inimigo com uma posicao 'x', 'y' e velocidade aleatorios    
    }
    #Adiciona 'novo_inimigo' a lista 'posicoes_inimigos'
    posicoes_inimigos.append(novo_inimigo)

#Loop continuará executando ate que a variavel seja verdadeira
#limita o jogo a rodar a 30 fps
#tela_blit e usado para desenhar a imagem de fundo no jogo
while not fim_de_jogo:
    
    relogio.tick(30)
    tela.blit(imagem_fundo, (0, 0))
    
    #Funcoes para verificar se as teclas apresentadas foram pressionadas
    for evento in pygame.event.get():
        
        if evento.type == pygame.QUIT:
            
            fim_de_jogo = True
            
        if evento.type == pygame.KEYDOWN:
                
            if evento.key == pygame.K_SPACE:
                    
                tiro_esquerda = {'x': posicao_jogador[0], 'y': posicao_jogador[1], 'velocidade': velocidade_tiro}
                tiros.append(tiro_esquerda)
                        
                tiro_direita = {'x': posicao_jogador[0] + tamanho_jogador, 'y':posicao_jogador[1], 'velocidade': velocidade_tiro}
                tiros.append(tiro_direita)
                                
                som_tiro.play()
                
                #Funcao acima foi utilizada para representar os tiros disparados do lado esquerdo e direito do jogador
                #Os tiros sao armazenados em uma lista (esquerda e direita) 
                #Adicionado o som do tiro toda vez que a tecla de espaco for pressionada
    
   #Funcao criada para utilizar todas as teclas e verificações das mesmas utilizadas 
    teclas = pygame.key.get_pressed()
    
    if teclas[pygame.K_LEFT] and posicao_jogador[0] > 0:
        posicao_jogador[0] -= velocidade_jogador
        
    if teclas[pygame.K_RIGHT] and posicao_jogador[0] < largura - tamanho_jogador:
        posicao_jogador[0] += velocidade_jogador
            
    if teclas[pygame.K_UP] and posicao_jogador[1] > 0:
        posicao_jogador[1] -= velocidade_jogador
            
    if teclas[pygame.K_DOWN] and posicao_jogador[1] < altura - tamanho_jogador:
        posicao_jogador[1] += velocidade_jogador
    
    #Loop 'for' para iterar sobre uma copia da lista 'posicoes_inimigos'
    #Aumenta a posicao 'y' do inimigo pela 'velocidade' do mesmo (simula o movimento)
    #Cria novos inimigos a partir da eliminacao de outros
    for inimigo in posicoes_inimigos[:]:
        inimigo['y'] += inimigo['velocidade']
        
        if inimigo['y'] > altura:
            posicoes_inimigos.remove(inimigo)
            criar_novo_inimigo()
            
    tiros_a_remover = []
    inimigos_a_remover = []
    
    #Loop 'for' usado para percorrer cada tiro na lista
    #Move o tiro para cima da tela
    #Se ele sair da tela, e adicionado a lista 'tiros_a_remover'
    for tiro in tiros:
        
        tiro['y'] -= tiro['velocidade']
        
        if tiro['y'] < 0:
            tiros_a_remover.append(tiro)
        
        #Loop 'for' usado para verificar se os tiros colidiram com o inimigo
        for tiro in tiros:
            for inimigo in posicoes_inimigos:
            
                if (inimigo['x'] < tiro['x'] < inimigo['x'] + tamanho_inimigo) and (inimigo['y'] < tiro['y'] < inimigo['y'] + tamanho_inimigo):
                        
                    tiros_a_remover.append(tiro)
                    inimigos_a_remover.append(inimigo)
                    explosoes.append({'x': inimigo['x'], 'y': inimigo['y'], 'tempo': pygame.time.get_ticks()})
            
                    pontos += 10
                
                    break
    
    #Loop 'for' que itera tiro na lista 'tiros_a_remover'
    for tiro in tiros_a_remover:
        
        if tiro in tiros:
            
            tiros.remove(tiro)
    
    #Loop 'for' que itera inimigo na lista 'inimigos_a_remover'
    for inimigo in inimigos_a_remover:
        
        if inimigo in posicoes_inimigos:
            
            posicoes_inimigos.remove(inimigo)
            
            criar_novo_inimigo()
            
    for tiro in tiros:
        
        pygame.draw.rect(tela, branco, (tiro['x'], tiro ['y'], 5, 10))
    
    tela.blit(jogador, (posicao_jogador[0], posicao_jogador[1])) #Desenha a imagem do jogador
    
    for inimigo in posicoes_inimigos:
        
        tela.blit(imagem_inimigo, (inimigo['x'], inimigo['y'])) #Desenha a imagem do inimigo
    
    texto_pontuacao = fonte.render(f"Pontos: {pontos}", True, branco) #Exibe a pontuacao do jogo na tela
    
    tela.blit(texto_pontuacao, (10, 10)) #Desenha a pontuacao na tela
    
    for inimigo in posicoes_inimigos:
        
        if(inimigo['x'] < posicao_jogador[0] < inimigo['x'] + tamanho_inimigo or
           inimigo['x'] < posicao_jogador[0] + tamanho_jogador < inimigo['x'] + tamanho_inimigo) and \
           (inimigo['y'] < posicao_jogador[1] < inimigo['y'] + tamanho_inimigo or
           inimigo['y'] < posicao_jogador[1] + tamanho_jogador < inimigo['y'] + tamanho_inimigo):
            
            vidas -= 1
            
            if vidas == 0:
                
                fim_de_jogo = True
                
            else:
                
                posicao_jogador = [largura // 2, altura - 2 * tamanho_jogador]
                
            break
            
            
    tempo_atual = pygame.time.get_ticks()
    
    explosoes_a_remover = [] #Cria uma lista para remover as explosoes
    
    #Loop 'for' criado para verificar se as explosoes ultrapassaram 500 milissegundos 
    #Se for True, ela e removida, se False, ela ainda sera exibida
    for explosao in explosoes:
        
        if tempo_atual - explosao['tempo'] > 500:
            
            explosoes_a_remover.append(explosao)
        
        else:
            
            tela.blit(imagem_explosao, (explosao['x'], explosao['y']))
            
    #Loop 'for' que percorre todas as explosoes que foram marcadas para serem removidas   
    for explosao in explosoes_a_remover:
        
        explosoes.remove(explosao) #Remove cada explosao da lista principal
    
    #Exibe a pontuacao e as vidas do jogador
    texto_pontuacao = fonte.render(f"Pontos: {pontos}", True, branco)
    texto_vidas = fonte.render(f"Vidas: {vidas}", True, branco)
    
    tela.blit(texto_pontuacao, (10, 10))
    tela.blit(texto_vidas, (10, 40))
    
    pygame.display.flip()

pygame.quit()
```
